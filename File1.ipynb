```python
from pyarabic.araby import tokenize
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk import sent_tokenize
nltk.download('punkt')

NewsText_words = []
urdu_sentences = df['News Text']
for i, sent in enumerate(urdu_sentences):
    words = word_tokenize(sent)
    for word in words:
        NewsText_words.append(word)
    

NewsText_words
```

```python
import csv
with open('urduwords.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Word'])
    for word in NewsText_words:
        writer.writerow([word])
```
