### File 1

```python
import pandas as pd

def remove_joined_ending(news_text):
    joined_endings = ['ہواجاتا', 'ہوجاتی', 'ہوجاتا', 'ہوجاتے', 'ہوگیا', 'ہوگئی', 'ہوگئے','جائیگی']
    for ending in joined_endings:
        # Check if the news text contains the joined ending
        if ending in news_text:
            # Split the news text at the joined ending
            parts = news_text.split(ending)
            # Return the first part of the split text with the ending
            return parts[0] + ending

    # If no joined ending is found, return the original news text
    return news_text

# Counter for changes
change_count = 0

for i, sentence in enumerate(df['News Text']):
    new_text = remove_joined_ending(sentence)
    if new_text != sentence:
        change_count += 1
        df.loc[i, 'News Text'] = new_text


df.to_excel('Ent1.xlsx', index=False)
print("Number of changes made:", change_count)
```

```python
for sent in df['News Text']:
    print(sent,'\n')
```
